```python
import numpy as np
```

```python
class NeuronNetwork:

    def sigmoid(self, X):
        return 1 / (1 + np.exp(-X))

    def sigmoidDerivative(self, X):
        return X * (1 - X)

    def feedForward(self, input, weights, bias):
        Z = np.dot(input, weights) + bias
        return self.sigmoid(Z)

    def backPass(self, predicted, actual, input):
        error = -(actual - predicted)
        d_pred = self.sigmoidDerivative(predicted)
        # Reshape error and d_pred to (n_samples, 1)
        error = error.reshape(-1, 1)
        d_pred = d_pred.reshape(-1, 1)
        # Broadcast input (n_samples, 4) to element-wise multiply
        dl_dw = np.dot(input.T, error * d_pred)
        return dl_dw

    def update_weight(self, learning_rate, dl_dw, weights):
        return weights - learning_rate * dl_dw.flatten()
```

```python
nn_obj= NeuronNetwork()

predicted= nn_obj.feedForward(1,0.5,0)
dl_dw= nn_obj.backPass(predicted,0,1)
print(dl_dw)
nn_obj.update_weight(0.1,dl_dw,0.5)

```

```python
df['binary_target'] = (df['target'] == 0).astype(int)

# Separate input and target
X = df.drop(['target', 'binary_target'], axis=1).values
y = df['binary_target'].values
```

```python
# add random weights and bais
np.random.seed(42)
weights = np.random.rand(4)
bias = np.random.rand(1)
```

```python
nn_obj = NeuronNetwork()

predicted = nn_obj.feedForward(X, weights, bias)
dl_dw = nn_obj.backPass(predicted, y, X)

print("dl_dw:", dl_dw)
print("weights:", weights)
updated_weights = nn_obj.update_weight(0.1, dl_dw, weights)
print("Updated weights:", updated_weights)
```
